#  Now the data we took is only from one single page. we have to make a loop and get the data from the pages we have in the web... so we will make create dataframes for all the pages and append them with old dataframe to get the final data of 10312  movies ( the rows we have in the website)

```python
#  first we have to create empty dataframe
df= pd.DataFrame()
```

```python
df              
# this is empty as there is nothing there in df. Now to this df we have to append all the other df's which we create for each page
```

```python
# in this website we have 500+ pages. here i would take only 200 pages for now. 
# so will run the loop for 200 times to fetch the data from each page 

for i in range(1, 516):
    response= requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=15b01f1783ed53eb29abacb337eb2254&language=en-US&page=1")
    temp_df= pd.DataFrame(response.json()['results'])[['id', 'title', 'original_language', 'release_date', 'overview', 'popularity', 'vote_average', 'vote_count']]
    df= df.append(temp_df, ignore_index= True)          

# setting ignore index = true so that everytime the data is fetched from the page the index number should 
# continue and not again repeat from 1 to 20 as there 20 movies in a page
```
